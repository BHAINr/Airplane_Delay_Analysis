```python
numeric_feature = ['arr_flights', 'arr_del15', 'carrier_ct', 'weather_ct', 'nas_ct',
                    'security_ct', 'late_aircraft_ct', 'arr_cancelled', 'arr_diverted',
                    'arr_delay', 'carrier_delay', 'weather_delay', 'nas_delay',
                    'security_delay', 'late_aircraft_delay']

for feature in numeric_feature:
    plt.figure(figsize=(8,5))
    sns.histplot(df[feature] , kde = True) 
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
categorical_feature = ['carrier' ,'carrier_name', 'airport', 'airport_name']

for feature in categorical_feature:
    top_categories = df[feature].value.counts().nlargest(15)
    
    plt.figure(figsize(12,6))
    sns.barplot(x = top_categories.index, y = top_categories.values)
    plt.title(f'top {15} {feature} count)
              
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()          
```

```python
df['date'] = pd.to_datetime(df[['year' , 'month']]).assign(day=1)
df.set_index('date' , inplace=True)

df['arr_flights'].fillna(df['arr_flights'].median(), inplace=True)

plt.figure(figsize=(20, 6))
plt.plot(df['arr_flights'], marker='o', linestyle='-')
plt.title('Trend of Arrival Flights Over Time')
plt.xlabel('Time')
plt.ylabel('Number of Arrival Flights')
plt.show()
```

```python
monthly_df = df.resample('M').sum()
plt.figure(figsize=(12, 6))
plt.plot(monthly_df['arr_flights'], marker='o', linestyle='-')
plt.title('Monthly Sum of Arrival Flights Over Time')
plt.xlabel('Time')
plt.ylabel('Number of Arrival Flights')
plt.show()
```

```python
# Yearly Aggregations
yearly_df = df.resample('Y').sum()
plt.figure(figsize=(12, 6))
plt.plot(yearly_df['arr_flights'], marker='o', linestyle='-')
plt.title('Yearly Sum of Arrival Flights Over Time')
plt.xlabel('Time')
plt.ylabel('Number of Arrival Flights')
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Assuming df has a DateTimeIndex or a column with datetime values
df = df.dropna(subset=['arr_flights'])

# Seasonal Decomposition
result = seasonal_decompose(df['arr_flights'], model='additive', period=12)

# Plot the decomposition
result.plot()
plt.show()
```
